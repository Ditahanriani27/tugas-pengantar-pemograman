```python
import numpy as np #mengimpor modul numpy untuk operasi matematika seperti: array, pi, dll

def integrate_rec(f, a, b, n): #mendefinisikan fungsi
    # Implementation of the rectanlge method 
    h = (b-a)/n #menghitung lebar tiap persegi panjang
    x = np.linspace(a, b, n+1) #membuat array titik dari a ke b sebanyak n+1 titik
    i=0 #inisialisasi variabel penghitung (index)
    area=0 #menyimpan total luas (awal=0)
    while i<n: #perulangan sebanyak n kali (untuk setiap kotak)
        sup_rect =  f(x[i])*h #menghitung luas persegi panjang
        area += sup_rect #menambahkan luas kotak ke total luas
        i += 1 #naik ke kotak berikutnya
    return area #mengembalikan hasil total luas
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_8 = integrate_rec(np.sin, 0, np.pi, 8) 
sup_20 = integrate_rec(np.sin, 0, np.pi, 20) 
sup_200 = integrate_rec(np.sin, 0, np.pi, 200)
#menghitung integral sin(x) dengan 8, 20, 200 kotak

print('Using n=8, the rectangle method returns a value of {:.2f}'.format(sup_8))
print('Using n=20, the rectangle method returns a value of {:.2f}'.format(sup_20))
print('Using n=200, the rectangle method returns a value of {:.2f}'.format(sup_200))
#menampilkan hasil untuk n=8, n=20, n=200
     

```

```python
import numpy as np #mengimpor modul numpy untuk operasi numerik seperti: array, pi, dll

def integrate_trap(f, a, b, n): #mendefinisikan fungsi
    # Implementation of the trapezoidal rule
    h = (b-a)/n #menghitung lebar tiap interval
    x = np.linspace(a, b, n+1) #membuat array dari a ke b sebanyak n+1 titik
    i=1 #mulai perhitungan dari indeks ke-1
    area = h*(f(x[0]) + f(x[n]))/2 #menghitung luas awal
    while i<n: #perulangan untuk menjumlahkan semua titik tengah
        sup_rect = f(x[i])*h #menghitung luas persegi kecil (tiap potongan kecil)
        area += sup_rect #menambahkan hasil ke total luas 
        i += 1 #naikkan indeks (ketitik berikutnya)
    return area #mengembalikan hasil integral pendekatan

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_8 = integrate_trap(np.sin, 0, np.pi, 8) #menghitung integral fungsi sin(x) dari 0 sampai pi dengan 8 bagian
sup_20 = integrate_trap(np.sin, 0, np.pi, 20)

print('Using n=8, the trapezoidal rule returns a value of {:.2f}'.format(sup_8))
print('Using n=20, the trapezoidal rule returns a value of {:.2f}'.format(sup_20))
   #menampilkan hasil
```

```python
import numpy as np #mengimpor modul numpy untuk operasi numerik 
from scipy import integrate #mengimpor modul integrasi numerik dari Scipy


x = np.linspace(0, 8, 4) # 3 divisions %*\textcolor{codegreen}{$[x_0,x_1,x_2]$}*), n=2 #membuat data x
y = x**2 #membuat fungsi y

sup_trapz = integrate.trapezoid(y,x) #menghitung luas dibawah kurva pakai metode trapesium
sup_simps = integrate.simpson(y,x) #menghitung luas pakai metode simpson
selisih = abs(sup_trapz - sup_simps) #menghitung selisih dari kedua metode


print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz))
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps))
#menampilkan hasil metode
print(f"selisih hasil kedua metode: {selisih:.2f}")
#menampilkan selisih kedua metode
```

```python
import numpy as np # mengimpor modul numpy untuk membuat array dan operasi matematika
from scipy import integrate #mengimpor modul integrasi numerik

conturs_areas = np.array([180000, 145000, 110000, 80000, 55000, 30000, 12000]) #menyimpan data luas penampang/kontur
x = np.array([0,20,40,60,80,100,120]) #menyimpan posisi (jarak atau kedalaman)

vol_traps = integrate.trapezoid(conturs_areas, x) #menghitung volume metode trapesium
vol_simps = integrate.simpson(conturs_areas, x) #menghitung volume metode simpson
selisih = abs(vol_traps - vol_simps) #menghitung selisih kedua metode

print('The trapezoidal rule returns a volume of {:.0f} cubic meters'.format(vol_traps))
print('The composite Simpson rule returns a volume of {:.0f} cubic meters'.format(vol_simps))
print(f"selisih kedua metode: {selisih:.0f} cubic meters")
#menampilkan hasil
```

```python
import numpy as np #mengimpor modul numpy untuk membuat array dan operasi numerik
from scipy.integrate import trapezoid #mengimpor fungsi trapezoid untuk integrasi numerik
import matplotlib.pyplot as plt #mengimpor matplotlib untuk membuat grafik

r = np.linspace(1,6400,6400) #membuat data jarak dari pusat bumi

def density(): #mendefinisikan fungsi untuk membuat data densitas bumi
    ro_inner_core = np.linspace(14500, 14000, 1200) #densitas inti dalam
    ro_outer_core = np.linspace(13000, 10500, 2200) #densitas inti luar
    ro_lower_mantle = np.linspace(6000,4800,2100) #densital mantel bawah
    ro_upper_mantle = np.linspace(4500,3600,750) #densitas mantel atas
    ro_crust = np.linspace(3100,2600,150) #densitas kerak bumi
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust)) #menggabungkan semua data jadi satu array dari pusat ke permukaan
    
    return ro_final #mengirim data densitas keluar dari fungsi

ro = density() #menjalankan fungsi

fig, ax = plt.subplots() #membuat area untuk plotting grafik
ax.plot(r,ro, label="Densities on the Earth's Interior", color='red') #menggambar grafik
ax.set_ylabel(r"Density [Kg/m$^3$]") #memberi label sumbu Y (densitas)
ax.set_xlabel("Distance from the Earth center r [km]") #memberi label sumbu X (jarak dari pusat bumi)
ax.legend() #menampilkan keterangan grafik
ax.grid() #menampilkan garis bantu agar mudah dibaca
```

```python
def gravity(r): #membuat fungsi bernama gravity
    
    g = np.zeros(len(r)) #membuat array berisi nol untuk menyimpan hasil gravitasi di tiap titik
    Gr = 6.673e-11 #nilai konstanta gravitasi
    r = r * 1000 # from Km to m #konversi satuan 
    
    for i in range(1,len(r)): #loop dari indeks 1 sampai akhir, menghitung gravitasi di tiap titik
        
        r1 = r[0:i] #jarak dari pusat sampai titik ke-i
        ro1  = ro[0:i] #densitas sampai titik ke-i
        r2 = r1[i-1] #mengambil nilai radius saat ini (terakhir)
    
        y = ro1*r1**2 #menghitung fungsi untuk integrasi
        y_int = trapezoid(y,r1) #menghitung integral menggunakan metode trapesium
        
        g1 = ((4 * np.pi*Gr)/(r2**2)) * y_int #menghitung gravitasi
        g[i] = g1 #simpan hasil ke array
           
    return g #mengembalikan semua nilai gravitasi

g = gravity(r) #menjalankan fungsi, menyimpan hasil ke variabel g

fig, ax = plt.subplots() #membuat area untuk plotting grafik
ax.plot(r,g, color='red') #menampilkan grafik
ax.grid() #menampilkan garis bantu
ax.set_ylabel(r'g [m/s$^2]$') #memberi label sumbu Y
ax.set_xlabel('Distance from the Earth center r [km]') #memberi label sumbu X
```

```python
def pressure(r, ro, g): #membuat fungsi untuk menghitung tekanan (jarak, densitas, gravitasi)
    
    p = np.zeros(len(r)) #menyimpan array kosong untuk hasil tekanan
    r = r *1000 #menghitung (konversi) km ke meter
    
    for i in range(0,len(r)): #perulangan untuk hitung tekanan di setiap titik  
        r1 = r[i:len(r)] #menyimpan potongan data jarak dari titik i ke luar bumi
        ro1 = ro[i:len(r)] #menyimpan densitas dari titik i ke luar
        g1 = g[i:len(r)] #menyimpan gravitasi
        y = ro1*g1 #menghitung hasil perkalian densitas x gravitasi
        p1 = trapezoid(y,r1) #menghitung tekanan dengan metode trapesium
        p[i] = p1 #menyimpan hasil tekanan ke array 
    return p #mengembalikan hasil tekanan
    
p = pressure(r,ro,g)/1e9 # expressed in GPa #pemanggil fungsi
z = np.linspace(6400, 1, 6400) #menyimpan data kedalaman dari permukaan ke pusat bumi

fig, ax = plt.subplots() #membuat area grafik
ax.plot(z,p, color='red') #menampilkan grafik
ax.grid() #menambahkan garis bantu digrafik
ax.set_ylabel('P [GPa]') #memberi label sumbu Y
ax.set_xlabel('Depth z from the Earth Surface [km]') #memberi label sumbu X
```
