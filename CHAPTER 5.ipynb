```python
import pandas as pd #untuk membaca dan mengolah data tabel (Excel)
from scipy.stats.mstats import gmean, hmean #untuk menghitung geometric mean dan harmonic mean
import matplotlib.pyplot as plt #untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca file Excel

a_mean = my_dataset.Rb.mean() #menghitung rata-rata aritmatika (mean biasa) dari kolom Rb
g_mean = gmean(my_dataset['Rb']) #menghitung geometric mean dari kolom Rb 
h_mean = hmean(my_dataset['Rb']) #menghitung harmonic mean dari kolom Rb

print ('-------') #menampilkan teks
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean)) #dibulatkan 1 angka dibelakang koma, #satuan (parts per million)
print ('-------')
#menampilkan hasil aritmetic mean

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')
#menampilkan hasil geometric mean

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')
#menampilkan hasil harmonic mean

fig, ax = plt.subplots() #membuat figure (area grafik)
ax.hist(my_dataset.Rb, bins='auto', density=True, edgecolor='k', color='gray', label='Measurements Hist', alpha=0.6) #membuat histogram dari kolom Rb
ax.axvline(a_mean, color='red', linestyle='-', label='Arithmetic mean', linewidth=1) #menggambar atau membuat garis vertikal arithmetic mean
ax.axvline(g_mean, color='orange', linestyle='-', label='Geometric mean', linewidth=1) #garis untuk geometric mean
ax.axvline(h_mean, color='green', linestyle='-', label='Harmonic mean', linewidth=1) #garis untuk harmonic mean
ax.set_xlabel(' [Rb ppm]') #memberi label sumbu
ax.set_ylabel('Probability density')
ax.legend() #menampilkan legenda
plt.show() #menampilkan grafik

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd #membaca dan mengolah data (Excel)
import matplotlib.pyplot as plt #membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca file Excel

variance =  my_dataset['Rb'].var() #menghitung varians dari data Rb
stddev =  my_dataset['Rb'].std() #menghitung simpanan baku (standar deviasi)

print ('-------')
print ('Variance')
print("{0:.0f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev))
print ('-------')
#menampilkan hasil perhitungan

fig, ax = plt.subplots() #membuat kanvas dan area plot
ax.hist(my_dataset.Rb, bins= 25, density = True, edgecolor='k', color='pink', label='Measurements Hist') #membuat histogram dari data kolom Rb
ax.axvline(my_dataset['Rb'].mean()-stddev, color='blue', label=r'mean - 1$\sigma$', linewidth=2) #membuat garis vertikal 
ax.axvline(my_dataset['Rb'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2)
ax.axvspan(my_dataset['Rb'].mean()-stddev, my_dataset['Rb'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') #membuat warna area pada dua nilai 
ax.set_xlabel('Rb [ppm]') #memberi nama label pada sumbu
ax.set_ylabel('Probability density')
ax.legend() #menampilkan legenda

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd #membaca dan mengolah data Excel
import numpy as np #untuk perhitungan matematika dan  statistik (persentil)
import matplotlib.pyplot as plt #membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca file data Excel

q1 = np.percentile(my_dataset.Rb, 25,) #menghitung kuartil 1
q3 = np.percentile(my_dataset.Rb, 75,) #menghitung kuartil 3
  
iqr = q3 - q1 # Interquaritle range (IQR) #menghitung jarak antara q1 dan q3

print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')
#menampilkan nilai IQR

fig, ax = plt.subplots() #membuat grafik area plot
ax.hist(my_dataset.Rb, bins='auto', density=True, edgecolor='k', label='Measurements Hist') #menampilkan histogram data kolom Rb
ax.axvline(q1, color='purple', label='Q1', linewidth=2) #garis vertikal untuk kuartil 1
ax.axvline(q3, color='green', label='Q3', linewidth=2) #garis vertikal untuk kuartil 3
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') #memberi warna area antara q1 dan q3
ax.set_xlabel('Rb [ppm]') 
ax.set_ylabel('Probability density')
#memberi nama sumbu X dan Y
ax.legend() #menampilkan keterangan grafik

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import numpy as np #menampilkan dan mengolah data Excel
from scipy.stats import skew #menghitung skewness (kemencengan data)

a_mean = my_dataset.Rb.mean() #menghitung rata-rata dari kolom Rb
median = my_dataset.Rb.median() #menghitung nilai tengah dari kolom Rb
hist, bin_edges = np.histogram(my_dataset['Rb'], bins=20, density=True) #menghitung histogram untuk modus, membagi data menjadi 20 kelompok
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 #mencari batang tertinggi dan modus
standard_deviation = my_dataset['Rb'].std() #menghitung penyebaran data dari rata-rata

a1 = (a_mean - modal_value) / standard_deviation  #menghitung skewness berdasarkan mean dan modus 
a2 = 3 * (a_mean - median) / standard_deviation #menghitung skewness berdasarkan mean dan median
g1 = skew(my_dataset['Rb']) #menghitung skewness otomatis(metode fisher-pearson)

print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')
#menampilkan nilai skewness dengan 2 angka desimal

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
